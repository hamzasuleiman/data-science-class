```python
df2=df[df["period"]== "Annual"]
df2
```

```python
plt.plot(df2["year"],df2["Agriculture"],ls="--",lw=3.5,marker="o",markevery=3,mec="red",mfc="green")
plt.xlabel("year", color= "#94035c", fontsize= 20)
plt.ylabel("Agriculturer value", color= "#833da8", fontsize= 20)
plt.title("The trend of agriculture over the year", color= "#12806d", fontsize= 20)
plt.style.use("ggplot")
```

```python
plt.style.available
```

```python
fig= plt.figure(figsize=(10,10))
axes = fig.add_axes([0.1,0.1,0.7,0.4])
plt.style.use("ggplot")
axes.plot(df2["year"],df2["Agriculture"],ls="--",lw=3.5,marker="o",markevery=3,mec="green",mfc="green",color="green",label="agric")
axes.plot(df2["year"],df2["Education"],ls="--",lw=3.5,marker="o",markevery=3,mec="blue",mfc="green",color="blue",label="edu")
axes.plot(df2["year"],df2["CrudepetroleumAndNaturalGas"],ls="--",lw=3.5,marker="o",markevery=3,mec="black",mfc="green",color="black",label="CRUD")
axes.plot(df2["year"],df2["InformationAndCommunication"],ls="--",lw=3.5,marker="o",markevery=3,mec="yellow",mfc="green",color="yellow",label="ICT")
axes.set_xlabel("YEAR", color= "#94035c", fontsize= 25)
axes.set_ylabel("GDP", color= "#94035c", fontsize= 25)
plt.title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color= "navy", fontsize= 20)
plt.legend()
plt.show()
```

```python
fig, axes= plt.subplots(2,2,figsize=(15,10))
axes[0,0].bar(df2["year"],df2["Agriculture"],color="green")
axes[0,0].set_xlabel("YEAR", color= "#94035c", fontsize= 12)
axes[0,0].set_ylabel("Agriculture", color= "navy", fontsize= 12)
axes[0,0].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color= "black", fontsize= 12)

axes[1,0].step(df2["year"],df2["Education"],color="purple")
axes[1,0].set_xlabel("YEAR", color= "#94035c", fontsize= 12)
axes[1,0].set_ylabel("Education", color= "navy", fontsize= 12)
axes[1,0].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color= "black", fontsize= 12)


axes[0,1].scatter(df2["year"],df2["CrudepetroleumAndNaturalGas"],color="blue")
axes[0,1].set_xlabel("YEAR", color= "#94035c", fontsize= 12)
axes[0,1].set_ylabel("CrudepetroleumAndNaturalGas", color= "navy", fontsize= 12)
axes[0,1].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color= "black", fontsize= 12)

axes[1,1].hist(df2["InformationAndCommunication"],bins="auto", color="red")
axes[1,1].set_xlabel("YEAR", color= "#94035c", fontsize= 12)
axes[1,1].set_ylabel("InformationAndCommunication", color= "navy", fontsize= 12)
axes[1,1].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color= "black", fontsize= 12)

plt.show()
```
