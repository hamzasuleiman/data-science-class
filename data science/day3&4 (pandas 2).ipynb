```python
df1=pd.read_csv("sample_pivot.csv")
df1.head(6)
```

```python
df1.isnull().sum()
```

```python
## copy the file to another variables
df3=df1.copy()
df3["Units"].fillna(np.mean(df3["Units"]),inplace=True)
df3.tail()
```

```python
df3.isnull().sum()
```

```python
## pivot table
df_pv=pd.pivot_table(df3,index="Region",columns="Type",values=["Sales"],aggfunc=min)
result=df_pv.head(6)
result
```

```python
## CROSSTAB
pd.crosstab(df3["Region"],df3["Type"],margins=True,margins_name="TOTAL")
```

```python
## MELT
df_melt=df_pv.melt(ignore_index=False)
df_melt
```

```python
df_melt.rename ({None:"Sales"},axis=1,inplace=True)
df_melt
```

```python
## what is the total sales for each region?
## which product type (type) has the highest average units sold per transaction?
## which product type has the highest total sale in each region?
```

```python
## what is the total sales for each region?
total_sales=df3.groupby("Region")["Sales"].sum()
total_sales
```

```python
## which product type (type) has the highest average units sold per transaction?
highest_avg=df3.groupby("Type")["Sales"].mean()
highest_avg
```

```python
## which product type has the highest total sale in each region?
highest_type=df3.groupby("Type")["Sales"].max()
highest_sort=highest_type.sort_values(ascending= False)
highest_sort
```
