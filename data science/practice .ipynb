```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("RealGDP16062023.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
df.rename({'index':'year', 'tyear':'period', 'Period':'Agriculture', 'Agriculture':'CropProduction', 'CropProduction':'Liverstock',
       'Livestock':'Forestry', 'Forestry':'Fishing', 'Fishing':'Industry', 'Industry':'MiningAndQuarrying', 'MiningAndQuarrying':'CrudepetroleumAndNaturalGas',
       'CrudePetroleumAndNaturalGas':'CoalMining', 'CoalMining':'MetalOres', 'MetalOres':'QuarryingAndotherMinerals',
       'QuarryingAndOtherMinerals':'Manufacturing', 'Manufacturing':'OilRefining', 'OilRefining':'Cement', 'Cement':'FoodBeverageAndTobacco',
       'FoodBeverageAndTobacco':'TextileAppareAndFootwear', 'TextileApparelAndFootwear':'WoodAndWoodProducts',
       'WoodAndWoodProducts':'pulpPaperAndPaperProducts', 'PulpPaperAndPaperProducts':'ChemicalAndPharmaceuticalProduct',
       'ChemicalAndPharmaceuticalProducts':'NonMetallicProducts', 'NonMetallicProducts':'PlasticAndRubberProducts',
       'PlasticAndRubberProducts':'ElectricalAndElectronics', 'ElectricalAndElectronics':'BasicMetalIronAndSteel',
       'BasicMetalIronAndSteel':'MotorVechiclesAndAssembly', 'MotorVehiclesAndAssembly':'OtherManufacturing',
       'OtherManufacturing':'ElectricityGasSteamAndAirCon', 'ElectricityGasSteamAndAirCon':'WaterSupplySewageWaste',
       'WaterSupplySewageWaste':'Construction', 'Construction':'Services', 'Services':'Trade', 'Trade':'AccommodationAndFoodServices',
       'AccommodationAndFoodServices':'TransportationAndStorage', 'TransportationAndStorage':'RoadTransport',
       'RoadTransport':'RailTransportAndPipelines', 'RailTransportAndPipelines':'WaterTransport', 'WaterTransport':'AirTransport',
       'AirTransport':'TransportServices', 'TransportServices':'PostAndCourierServices', 'PostAndCourierServices':'InformationAndCommunication',
       'InformationAndCommunication':'Publishing', 'Publishing':'MotionPicturesSoundRecordingAndMusicProduction',
       'MotionPicturesSoundRecordingAndMusicProduction':'Broadcasting', 'Broadcasting':'ArtsEntertainmentAndRecreation',
       'ArtsEntertainmentAndRecreation':'FinanceAndInsurance', 'FinanceAndInsurance':'FinancialInstitutions',
       'FinancialInstitutions':'Insurance', 'Insurance':'RealEstate', 'RealEstate':'ProfessionalScientificAndTechnicalServices',
       'ProfessionalScientificAndTechnicalServices':'AdministrativeAndSupportServicesBusinessServices',
       'AdministrativeAndSupportServicesBusinessServices':'PublicAdministration',
       'PublicAdministration':'Education', 'Education':'HumanHealthAndSocialServices', 'HumanHealthAndSocialServices':'OtherServices',
       'OtherServices':'GDP at 2010 Constant Basic Prices', 'GDP at 2010 Constant Basic Prices':'NetTaxesOnProducts',
       'NetTaxesOnProducts':'NetTaxesOnProducts', 'GDP at 2010 Constant Market Prices':'NAN'},axis = 1,inplace=True)
df.head()
```

```python
df.drop("NAN",axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
fig, axes= plt.subplots(2,2,figsize=(15,10))
axes[0,0].bar(df["year"],df["CropProduction"],color="green")
axes[0,0].set_xlabel("YEAR", color= "#94035c", fontsize= 12)
axes[0,0].set_ylabel("CropProduction", color= "navy", fontsize= 12)
axes[0,0].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color= "black", fontsize= 12)

axes[1,0].step(df["year"],df["Fishing"],color="purple")
axes[1,0].set_xlabel("YEAR", color= "#94035c", fontsize= 12)
axes[1,0].set_ylabel("Fishing", color= "navy", fontsize= 12)
axes[1,0].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color= "black", fontsize= 12)


axes[0,1].scatter(df["year"],df["RealEstate"],color="blue")
axes[0,1].set_xlabel("YEAR", color= "#94035c", fontsize= 12)
axes[0,1].set_ylabel("RealEstate", color= "navy", fontsize= 12)
axes[0,1].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color= "black", fontsize= 12)

axes[1,1].hist(df["PublicAdministration"],bins="auto", color="red")
axes[1,1].set_xlabel("YEAR", color= "#94035c", fontsize= 12)
axes[1,1].set_ylabel("PublicAdministration", color= "navy", fontsize= 12)
axes[1,1].set_title("TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS", color= "black", fontsize= 12)

plt.show()
```

```python
df2=df[["Liverstock","Forestry","Industry","MiningAndQuarrying"]]
df2.loc["Total"]= (sum(df2["Liverstock"]), sum(df2["Forestry"]), sum(df2["Industry"]),sum(df2["MiningAndQuarrying"]))
df2.tail()
```

```python
fig, axes=plt.subplots(figsize=(6,6),dpi=120)
axes.pie(df2.loc["Total"],labels=df2.columns,autopct= "%1.2f%%",shadow= True, explode =[0.0,0.0,0.0,0.0],colors=["white","blue","green","black"])
plt.show()
```
